**QuantileTransformer**
  * non linear transformation
  * smooths out unusual distributions and is less influenced by outliers than scaling methods
  * to map data to a normal distribution set output_distribution='normal'

*Formula*
G inverse(F(x))
  F is cumulative distribution function
  G quantile function of the desired output distribution


[QuantileTransformer](https://scikit-learn.org/stable/modules/preprocessing.html#preprocessing)

```python
from sklearn.datasets import load_iris
from sklearn.preprocessing import QuantileTransformer
```

```python
X, y = load_iris(return_X_y=True)
quantile_transformer = QuantileTransformer(random_state=0)
quantile_transformer.fit_transform(X)
```

```python
quantile_transformer = QuantileTransformer(output_distribution='normal', random_state=0)
quantile_transformer.fit_transform(X)
```
