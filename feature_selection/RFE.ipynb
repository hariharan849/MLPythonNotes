**Recursive Feature Elimination**
  * recursively eliminates features based on coef_ or feature_importance attribute

```python
from sklearn.feature_selection import RFE
from sklearn.datasets import load_digits
from sklearn.linear_model import LogisticRegression
```

```python
X, y = load_digits(return_X_y=True)
X.shape
```

```python
model = LogisticRegression()
feature = RFE(model, 3)
feature.fit(X, y)
```

```python
print ("No of feature: {0}".format(feature.n_features_))
print ("Selected feature: {0}".format(feature.support_))
print ("Feature ranking: {0}".format(feature.ranking_))
```
